```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt', delim_whitespace=True)
```

```python
# Compare the average income based on ethnicity

print(df.groupby(['Ethnicity'])["Income"].mean())
```

```python
# On average single people have a slightly higher Balanced compared to married people
print(df.groupby(['Married'])["Balance"].mean())
```

```python
# Highest income
df["Income"].max()
```

```python
#Lowest income
df["Income"].min()
```

```python
# How many cards do we have recorded in our dataset? (Hint: use `sum()`)
df["Cards"].sum()
```

```python
# Number of females compared to males 
print(df.groupby(['Gender'])["Gender"].count())
```
